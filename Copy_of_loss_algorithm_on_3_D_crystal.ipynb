```
import matplotlib.pyplot as plt
import numpy as np

# 3D crystal
def draw_3d_crystal():
    x, y, z = np.random.randint(0, 100, 300)
    x, y = x[:, None], y[:, None]
    plt.scatter(x, y, z, c="blue")
    plt.title("3D Crystal")

# Loss algorithm
def apply_loss(x, y, z):
    for i in range(len(x)):
        if x[i] > 50:
            x[i] = 0
        elif y[i] > 50:
            y[i] = 0
        elif z[i] > 50:
            z[i] = 0
    return x, y, z

# Apply the loss algorithm
x, y, z = apply_loss(x, y, z)

# Plot the crystal
plt.scatter(x, y, z, c="blue")
plt.title("3D Crystal with Loss")
plt.show()

# Call the functions to draw the crystals
draw_3d_crystal()
apply_loss(x, y, z)
```
