```
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes to the graph
G.add_nodes_from([
    "Light", "Crystal Properties", "Refractive Index Coefficient",
    "Increased Scattering Cross Sections", "Increased Absorption Cross Sections",
    "Electromagnetic Field Intensity", "Absorbed Light", "Scattered Light",
    "Heat Generated", "Temperature Rise", "Change in Refractive Index",
    "Change in Direction of Incident Light", "Transmitted Light",
    "Light Absorbed by Scattering"
])

# Add edges to the graph
G.add_edges_from([
    ("Light", "Crystal Properties"),
    ("Crystal Properties", "Refractive Index Coefficient"),
    ("Crystal Properties", "Increased Scattering Cross Sections"),
    ("Increased Scattering Cross Sections", "Increased Absorption Cross Sections"),
    ("Increased Absorption Cross Sections", "Electromagnetic Field Intensity"),
    ("Electromagnetic Field Intensity", "Absorbed Light"),
    ("Absorbed Light", "Heat Generated"),
    ("Increased Scattering Cross Sections", "Scattered Light"),
    ("Heat Generated", "Temperature Rise"),
    ("Refractive Index Coefficient", "Change in Refractive Index"),
    ("Change in Refractive Index", "Change in Direction of Incident Light"),
    ("Change in Direction of Incident Light", "Transmitted Light"),
    ("Electromagnetic Field Intensity", "Transmitted Light"),
    ("Increased Scattering Cross Sections", "Light Absorbed by Scattering")
])

# Set node positions for better layout
pos = nx.spring_layout(G)

# Adjust edge node positions
for edge in G.edges():
    source, target = edge
    x_diff = pos[target][0] - pos[source][0]
    y_diff = pos[target][1] - pos[source][1]
    if x_diff > 0.1:
        pos[target][0] -= 0.1
    elif x_diff < -0.1:
        pos[target][0] += 0.1
    if y_diff > 0.1:
        pos[target][1] -= 0.1
    elif y_diff < -0.1:
        pos[target][1] += 0.1

# Draw the graph
plt.figure(figsize=(15, 8))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=1000,
    node_color="red",
    edge_color="gray",
    arrowsize=20,
    font_size=14
)
plt.title("Flowchart: Algorithm Steps")
plt.axis("off")
plt.show()
```
