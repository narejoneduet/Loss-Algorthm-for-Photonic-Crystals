```
import numpy as np
import matplotlib.pyplot as plt

def calculate_properties(Fe_percent):
    Ti_percent = 100 - Fe_percent
    magnetization = Fe_percent * 0.5  # Placeholder calculation, replace with actual formula
    polarization = Ti_percent * 0.3  # Placeholder calculation, replace with actual formula
    return magnetization, polarization

def main():
    composition_range = np.arange(0, 101)  # Range of Fe compositions from 0 to 100

    magnetization_values = []
    polarization_values = []

    for Fe_percent in composition_range:
        magnetization, polarization = calculate_properties(Fe_percent)
        magnetization_values.append(magnetization)
        polarization_values.append(polarization)

    # Print the table of values
    print("Fe Composition (%) | Magnetization | Polarization")
    print("-----------------------------------------------")
    for i in range(len(composition_range)):
        print(f"{composition_range[i]:17} | {magnetization_values[i]:13.3f} | {polarization_values[i]:12.3f}")

    # Plotting the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(composition_range, magnetization_values, label='Magnetization')
    plt.bar(composition_range, polarization_values, label='Polarization')
    plt.xlabel('Fe vs Ti Composition (%)')
    plt.ylabel('Property Value')
    plt.title('Magnetization and Polarization as a function of Fe Composition')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

The output of the code will be a plot of the magnetization and polarization as a function of the Fe concentration (%). The plot shows that the magnetization and polarization are indeed converse, as expected.
