```
import numpy as np
import matplotlib.pyplot as plt

# Define the lattice parameters
a = 1.0
b = 1.0

# Define the refractive indices of the materials
n1 = 1.5
n2 = 4.0

# Calculate the band structure
k = np.linspace(-2 * np.pi, 2 * np.pi, 100)
bands = np.zeros((len(k), 2))
for i in range(len(k)):
    bands[i, 0] = np.sqrt(n1**2 + k[i]**2)
    bands[i, 1] = np.sqrt(n2**2 + k[i]**2)

# Plot the dispersion plot
plt.plot(k, bands[:, 0], label="TM")
plt.plot(k, bands[:, 1], label="TE")
plt.xlabel("Wavevector")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
