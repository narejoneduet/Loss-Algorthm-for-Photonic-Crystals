```
import numpy as np
import matplotlib.pyplot as plt

def calculate_effective_mass(k):
    """
    Calculates the effective mass for a given wavevector.

    Args:
        k: The wavevector.

    Returns:
        The effective mass.
    """
    # Define the parameters for the effective mass calculation
    m0 = 9.10938356e-31  # Electron rest mass in kg
    hbar = 1.0545718e-34  # Reduced Planck constant in J s
    a = 1  # Lattice constant in m

    # Calculate the effective mass
    effective_mass = hbar**2 * k**2 / (2 * m0 * a**2)

    return effective_mass

# Define the wavevector range
k = np.linspace(-np.pi, np.pi, 100)

# Calculate the effective mass for each wavevector
effective_mass = calculate_effective_mass(k)

# Plot the effective mass as a function of wavevector
plt.plot(k, effective_mass)
plt.xlabel('Wavevector (k)')
plt.ylabel('Effective Mass (m*)')
plt.title('Effective Mass vs Wavevector')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def fermi_dirac_distribution(energy, fermi_energy):
    """
    Calculates the Fermi-Dirac distribution function.

    Args:
        energy: The energy values.
        fermi_energy: The Fermi energy.

    Returns:
        The Fermi-Dirac distribution function.
    """
    return 1 / (1 + np.exp((energy - fermi_energy) / (0.0000861733)))  # Boltzmann constant k_B = 8.61733e-5 eV/K

# Define the energy range
energy = np.linspace(-1, 1, 100)

# Define the Fermi energy
fermi_energy = 0

# Calculate the Fermi-Dirac distribution function for each energy value
fermi_dirac = fermi_dirac_distribution(energy, fermi_energy)

# Plot the Fermi-Dirac distribution function
plt.plot(energy, fermi_dirac)
plt.xlabel('Energy (eV)')
plt.ylabel('Fermi-Dirac Probability')
plt.title('Fermi-Dirac Distribution Function at T = 0 K')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def fermi_dirac_distribution(energy, fermi_energy, temperature):
    """
    Calculates the Fermi-Dirac distribution function.

    Args:
        energy: The energy values.
        fermi_energy: The Fermi energy.
        temperature: The temperature in Kelvin.

    Returns:
        The Fermi-Dirac distribution function.
    """
    k_B = 8.617333262145e-5  # Boltzmann constant in eV/K
    return 1 / (1 + np.exp((energy - fermi_energy) / (k_B * temperature)))

# Define the energy range
energy = np.linspace(-1, 1, 100)

# Define the Fermi energy
fermi_energy = 0

# Define the temperatures
temperatures = [0, 300, 1000]

# Plot the Fermi-Dirac distribution function for each temperature
for temperature in temperatures:
    fermi_dirac = fermi_dirac_distribution(energy, fermi_energy, temperature)
    plt.plot(energy, fermi_dirac, label=f'T = {temperature} K')

plt.xlabel('Energy (eV)')
plt.ylabel('Fermi-Dirac Probability')
plt.title('Fermi-Dirac Distribution Function')
plt.legend()
plt.show()
```
